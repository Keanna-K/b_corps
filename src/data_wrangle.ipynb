```python
import pandas as pd
import googlemaps
import json
```

```python
# Read in data set
# data from https://data.world/blab/b-corp-impact-data
bcorp = pd.read_csv('data/b_corp_data.csv')
```

```python
# Select relevant columns
bcorp = bcorp.iloc[:,0:22]
bcorp.head()
```

```python
# select only the Ontario corporations
ont_bcorp = bcorp[bcorp.state == 'Ontario']
ont_bcorp.reset_index(drop=True, inplace=True)
ont_bcorp
```

```python
with open('creds.json') as f:
    data = json.load(f)
api_key = data['key']
```

```python
gmaps = googlemaps.Client(key=api_key)
```

```python
# initialize empty list
geom = []
specific = []

# Geocoding the company addresses
for company in ont_bcorp.index:
    address = ont_bcorp.company_name[company]
    components = {'country': 'CA', 
                  'administrative_area': 'ON',
                  'locality': ont_bcorp.city[company]}
    geocode_result = gmaps.geocode(address, components=components)
    
    # extract long and lat from result
    try:
        geom.append(geocode_result[0]['geometry']['location'])
        specific.append(1)
    except:
        components = {'country': 'CA', 
                      'administrative_area': 'ON'}
        geocode_result = gmaps.geocode(ont_bcorp.city[company], components=components)
        geom.append(geocode_result[0]['geometry']['location'])
        specific.append(0)
```

```python
# add the latitude and longitude info to the dataframe
ont_bcorp['geom'] = geom
ont_bcorp['specific'] = specific
ont_bcorp.head()
```

```python
# Save dataframe to CSV file
ont_bcorp.to_csv('data/ont_bcorps.csv', index=False)
```

```python
# Get companies with non specific locations
missing_geom = ont_bcorp[ont_bcorp.specific == 0].drop_duplicates(subset=['company_name'])
missing_geom.company_name
```
